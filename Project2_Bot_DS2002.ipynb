```python
def unique(list1):
    # initialize a null list
    unique_list = []
    # traverse for all elements
    for x in list1:
        # check if exists in unique_list or not
        if x not in unique_list:
            unique_list.append(x)
    return unique_list
```

```python
types = {
    "type1": ["best", "top", "highest rating", "most popular"],
    "type3": ["long", "length", "season"],
    "type4": ["release", "when"],
    "type5": ["rating", "score"],
    "type6": ["who", "play", "direct", "role"],
    "type7": ["genre"],
    "type8": ["where", "produce"],
    "type9": ["type", "is"],
    "type10": ["vote"]}
```

```python
def bot(question):
    
    global types, all_titles
    
    keys = []
    error_message = "Sorry. I don't understand your question.\nI am able to answer 10 types of questions.\n\nPlease ask me again in the following structures:\n\nType 1: Top shows/movies\nWhat were the top X shows/movies on Netflix X years ago?\nWhat were the top X shows/movies on Netflix last year?\n\nType2: Top show/movie in year XXXX\nWhat was the top/best show/movie on Netflix in XXXX?\nWhat was the top/best show/movie on Netflix last year?\n\nType 3: Lengh\nHow long was XXXX?\nHow many seasons does XXXX have?\n\nType 4: Release year\nWhen did XXXX release?\nWhen did XXXX take place?\n\nType 5: Rating \nWhat is the rating of XXXX?\nWhat is the score of XXXX?\n\nType 6: Cast\nWho played XXX in XXX?\nWhat's XXX's role in XXXX?\nWho did XXX play in XXXX?\nWho directed XXXX?\n\nType 7: Genre\nWhat is XXXX's genera?\n\nType 8: Production\nWhere was XXXX produced?\n\nType 9: Type\nWhat's XXX's type?\nIs XXX a show/movie?\nIs XXX a show or a movie?\n\nType 10: Vote\nHow many people voted for XXX?"
    try:
        # Find type  
        for i in types:
            for j in types[i]:
                if j in question:
                    keys.append(i)
        keys = unique(keys)
        keys = [keys[0]]

        # Type 1
        # keys = [type1, "movie/show", Top n, year]
        if "type1" in keys:
            if "show" in question:
                keys.append("show")
            elif "movie" in question:
                keys.append("movie")

            numbers = re.findall(r'\d+', question) 

        # Type 2
        # keys = [type2, "movie/show", year]
            if len(numbers) == 0:
                keys[0] = "type2"
                if "last" in question:
                    keys.append(int(2021))
            if len(numbers) == 1:
                if "last" in question:
                    keys[0] = "type1"
                    keys.append(numbers[0])
                    keys.append(int(2021))
                else: 
                    keys[0] = "type2"
                    keys.append(numbers[0])
            elif len(numbers) == 2:
                keys.append(numbers[0])
                if "this" in question:
                    keys.append("2022")
                elif len(numbers[1]) == 4:
                    keys.append(numbers[1])
                elif "last" in question: 
                    keys.append(int(2022) - 1)
                else:
                    keys.append(int(2022) - int(numbers[1]))  
            return keys

        # Type 3
        # keys = [type3, "show/movie", title]
        elif "type3" in keys:
            if "season" in question:
                keys.append("show")
            else:
                keys.append("movie")
            for i in all_titles:
                if i in question:
                    keys.append(i)
            if len(keys) == 2:
                return print("I don't have the movie/show you are looking for.")
            else: 
                target = input("Which one are you talking about? Choose from the followings(enter a number): " + str(keys[2:]).title())
                keys = [keys[0], keys[1], keys[int(target)+1]]
                return keys

        # Type 4
        # keys = [type4, "release_year", title]
        elif "type4" in keys: 
            keys.append("release_year")
            for i in all_titles: 
                if i in question:
                    keys.append(i)
            if len(keys) == 2:
                return print("I don't have the movie/show you are looking for.")
            else: 
                target = input("Which one are you talking about? Choose from the followings(enter a number): " + str(keys[2:]).title())
                keys = [keys[0], keys[1], keys[int(target)+1]]
                return keys

        # Type 5
        # keys = [type5, "imdb_score", title]
        elif "type5" in keys: 
            keys.append("imdb_socre")
            for i in all_titles: 
                if i in question:
                    keys.append(i)
            if len(keys) == 2:
                return print("I don't have the movie/show you are looking for.")
            else: 
                target = input("Which one are you talking about? Choose from the followings(enter a number): " + str(keys[2:]).title())
                keys = [keys[0], keys[1], keys[int(target)+1]]
                return keys

        # Type 6 
        # keys = [type6, title, "want_name/character/director", name/character]
        elif "type6" in keys:
            for i in all_titles: 
                if i in question:
                    keys.append(i)
            keys = keys[0:2]
            if "who played" in question:
                keys.append("want_name")
                pattern1 = r"played (\w+) in"
                match = re.search(pattern1, question)
                if match:
                    role = match.group(1)
                keys.append(str(role))
            if "what's" in question:
                keys.append("want_character")
                pattern2 = r"what's (\w+)'s role"
                match = re.search(pattern2, question)
                if match:
                    name = match.group(1)
                keys.append(str(name))
            if "who did" in question:
                keys.append("want_character")
                pattern3 = r"did (\w+) play"
                match = re.search(pattern3, question)
                if match:
                    name = match.group(1)
                keys.append(str(name))
            if "direct" in question:
                keys.append("want_director")    
            return keys          

        # Type 7
        # keys = [type7, title]
        elif "type7" in keys:
            for i in all_titles: 
                if i in question:
                    keys.append(i)
            if len(keys) == 1:
                return print("I don't have the movie/show you are looking for.")
            else: 
                target = input("Which one are you talking about? Choose from the followings(enter a number): " + str(keys[1:]).title())
                keys = [keys[0], keys[int(target)]]
                return keys

        # Type 8 
        # keys = [type8, title]
        elif "type8" in keys:
            for i in all_titles: 
                if i in question:
                    keys.append(i)
            if len(keys) == 1:
                return print("I don't have the movie/show you are looking for.")
            else: 
                target = input("Which one are you talking about? Choose from the followings(enter a number): " + str(keys[1:]).title())
                keys = [keys[0], keys[int(target)]]
                return keys
        # Type 9
        # keys = [type9, title]
        elif "type9" in keys:
            for i in all_titles: 
                if i in question:
                    keys.append(i)
            if len(keys) == 1:
                return print("I don't have the movie/show you are looking for.")
            else: 
                target = input("Which one are you talking about? Choose from the followings(enter a number): " + str(keys[1:]).title())
                keys = [keys[0], keys[int(target)]]
                return keys

        # Type 10
        # keys = [type10, title]
        elif "type10" in keys:
            for i in all_titles: 
                if i in question:
                    keys.append(i)
            if len(keys) == 1:
                return print("I don't have the movie/show you are looking for.")
            else: 
                target = input("Which one are you talking about? Choose from the followings(enter a number): " + str(keys[1:]).title())
                keys = [keys[0], keys[int(target)]]
                return keys
        else:
            return error_message
    except:
        return error_message
```

```python
questions = []
answers = []
keys = []
on = True
more_question = "yes"
def ask():
    global on, more_question
    
    question = input('')
    question = question.lower()
    questions.append(question)
    
    key = bot(question)
    keys.append(key)
    
    my_answer = answer(key)
    print(my_answer)
    answers.append(my_answer)
    
    more_question = input("Do you have more questions?(Yes, No): ")
    more_question = more_question.lower()
    if more_question == "no":
        on = False
    return answer
```
