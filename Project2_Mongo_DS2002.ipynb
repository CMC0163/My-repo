```python
best_shows = pd.read_csv("/Users/CMC/Desktop/Fall 2022/DS 2002/Project2/Best Shows Netflix.csv")
best_shows = best_shows.applymap(lambda x: x.lower() if isinstance(x, str) else x)
```

```python
best_show_by_year = pd.read_csv("/Users/CMC/Desktop/Fall 2022/DS 2002/Project2/Best Show by Year Netflix.csv")
best_show_by_year = best_show_by_year.applymap(lambda x: x.lower() if isinstance(x, str) else x)
```

```python
best_movies = pd.read_csv("/Users/CMC/Desktop/Fall 2022/DS 2002/Project2/Best Movies Netflix.csv")
best_movies = best_movies.applymap(lambda x: x.lower() if isinstance(x, str) else x)
```

```python
best_movie_by_year = pd.read_csv("/Users/CMC/Desktop/Fall 2022/DS 2002/Project2/Best Movie by Year Netflix.csv")
best_movie_by_year = best_movie_by_year.applymap(lambda x: x.lower() if isinstance(x, str) else x)
```

```python
titles = pd.read_csv("/Users/CMC/Desktop/Fall 2022/DS 2002/Project2/raw_titles.csv")
titles = titles.applymap(lambda x: x.lower() if isinstance(x, str) else x)
titles = titles.fillna(0)
```

```python
credits = pd.read_csv("/Users/CMC/Desktop/Fall 2022/DS 2002/Project2/raw_credits.csv")
credits = credits.applymap(lambda x: x.lower() if isinstance(x, str) else x)
```

```python
# Type 1: Top shows/movies
# What were the top 5 shows/movies on Netflix 2 years ago?
# What were the top 5 shows/movies on Netflix last year?

def type1(key):
    # keys = [type1, "movie/show", Top n, year]
    years = list(best_movies["RELEASE_YEAR"])
    if str(key[3]) in str(years):
        if key[1] == "movie":
            df = best_movies.sort_values("SCORE", ascending = False)
            df1 = df[df["RELEASE_YEAR"] == int(key[3])]
            top = list(df1.head(int(key[2]))["TITLE"])

        elif key[1] == "show":
            df = best_shows.sort_values("SCORE", ascending = False)
            df1 = df[df["RELEASE_YEAR"] == int(key[3])]
            top = list(df1.head(int(key[2]))["TITLE"])

        answer = "The top " + str(key[2]) + " " + str(key[1]) + "s in year " + str(key[3]) + " are " 

        return print(answer, str(top).title())
    else: 
        return print("No top shows or movies that year.")
```

```python
# Type 2: Top show/movie in year XXXX
# What was the top/best show/movie on Netflix in XXXX?
# What was the top/best show/movie on Netflix last year?

def type2(key):
    # keys = [type2, "movie/show", year]
    years = list(best_movie_by_year["RELEASE_YEAR"])
    if str(key[2]) in str(years):
        if key[1] == "movie":
            df = best_movie_by_year[best_movie_by_year["RELEASE_YEAR"] == int(key[2])]
            top = list(df["TITLE"])

        elif key[1] == "show":
            df = best_show_by_year[best_show_by_year["RELEASE_YEAR"] == int(key[2])]
            top = list(df["TITLE"])

        answer = "The best " + str(key[1]) + " in year " + str(key[2]) + " is " + str(top[0]).title() + " ."

        return print(answer)
    else: 
        return print("No top show or movie that year.")
```

```python
# Type 3: Lengh
# How long was XXXX?
# How many seasons does XXXX have?

def type3(key):
    # keys = [type3, "show/movie", title]
    all_my_titles = list(titles["title"])
    
    if key[1] == "movie":
        if str(key[2]) in str(all_my_titles):
            df = titles[titles["title"] == str(key[2])]
            length = list(df["runtime"])
            years = list(df["release_year"])
            if len(length) == 1:
                length = [length[0]]
            else: 
                try:
                    year = input("Which one are you asking about? Choose one of the following release years: " + str(years))
                    index = years.index(int(year))
                    length = [length[int(index)]]
                    answer = "The duration of the movie " + str(key[2].title()) + " is " + str(length[0]) + " minutes."
                    return answer
                except: 
                    print("Oops, the selected year is not found")
        else:
            return print("Can't find the movie you are asking about.")
        
    elif key[1] == "show":
        if str(key[2]) in str(all_my_titles):
            df = titles[titles["title"] == str(key[2])]
            length = list(df["seasons"])
            years = list(df["release_year"])
            if len(length) == 1:
                length = [length[0]]
            else: 
                try:
                    year = input("Which one are you asking about? Choose one of the following release years: " + str(years))
                    index = years.index(int(year))
                    length = [length[int(index)]]
                    answer = "The show " + str(key[2].title()) + " has " + str(length[0]) + " seasons."
                    return answer
                except: 
                    print("Oops, the selected year is not found")
        else: 
            return print("Can't find the show you are asking about.")
        
```

```python
# Type 4: Release year
# When did XXXX release?
# When did XXXX take place?

def type4(key):
    # keys = [type4, "release_year", title]
    all_my_titles = list(titles["title"])
    if str(key[2]) in str(all_my_titles):
        df = titles[titles["title"] == str(key[2])]
        year = list(df["release_year"])
        if len(year) == 1: 
            answer = "The release year of " + str(key[2].title()) + " is " + str(year[0]) + "."
        else: 
            answer = "There are " + str(len(year)) + " versions of " + str(key[2].title()) + ", and the release years are: " + str(year) + "."
        return answer
    else: 
        return print("Can't find the show you are asking about.")
```

```python
# Type 5: Rating 
# What is the rating of XXXX?
# What is the score of XXXX?

def type5(key):
    # keys = [type5, "imdb_score", title]
    df = titles[titles["title"] == str(key[2])]
    rating = list(df["imdb_score"])
    years = list(df["release_year"])
    if len(rating) == 1:
        rating = [rating[0]]
        if rating[0] == 0:
            return print("The rating for this movie/show is unavailable.")
    else: 
        try:
            year = input("Which one are you asking about? Choose one of the following release years: " + str(years))
            index = years.index(int(year))
            rating = [rating[int(index)]]
            if rating[0] == 0:
                return print("The rating for this movie/show is unavailable.")
            else: 
                answer = "The rating of " + str(key[2].title()) + " is " + str(rating[0]) + "."
                return answer
        except: 
            print("Oops, the selected year is not found")
```

```python
# Type 6: Cast
# Who played XXX in XXX?
# What's XXX's role in XXXX?
# Who did XXX play in XXXX?
# Who directed XXXX?

def type6(key):
    # keys = [type6, title, "want_name/character/director", name/character]
    df = titles[titles["title"] == str(key[1])]
    myid = list(df["id"])
    years = list(df["release_year"])
    if len(myid) == 1:
        myid = [myid[0]]
    else: 
        try:
            year = input("Which one are you asking about? Choose one of the following release years: " + str(years))
            index = years.index(int(year))
            myid = [myid[int(index)]]
        except: 
            print("Oops, the selected year is not found")
    df2 = credits[credits["id"] == str(myid[0])]
    if str(key[2]) == "want_name":
        try: 
            df3 = df2[df2["character"] == str(key[3])]
            name = list(df3["name"])
            answer = "In " + str(key[1]).title() + ", " + str(name[0]).title() + " played " + str(key[3]).title() + "."
            return answer
        except: 
            return print("Oops. Can't find the person who played the role.")
    if str(key[2]) == "want_character":
        try: 
            df3 = df2[df2["name"] == str(key[3])]
            character = list(df3["character"])
            answer = "In " + str(key[1]).title() + ", " + str(key[3]).title() + " played " + str(character[0]).title() + "."
            return answer
        except: 
            return print("Oops. Can't find which role the person played.")
    if str(key[2]) == "want_director":
        try: 
            df3 = df2[df2["role"] == "director"]
            director = list(df3["name"])
            answer = str(director[0]).title() + " directed " + str(key[1]).title() + "."
            return answer
        except: 
            return print("Oops. Can't find the director.")
```

```python
# Type 7: Genre
# What is XXXX's genre?

def type7(key):
    # keys = [type7, title]
    df = titles[titles["title"] == str(key[1])]
    genre = list(df["genres"])
    years = list(df["release_year"])
    if len(genre) == 1:
        genre = [genre[0]]
    else: 
        try:
            year = input("Which one are you asking about? Choose one of the following release years: " + str(years))
            index = years.index(int(year))
            genre = [genre[int(index)]]
            answer = "The genre of " + str(key[1].title()) + " is " + str(genre[0]) + "."
            return answer
        except: 
            print("Oops, the selected year is not found")
```

```python
# Type 8: Production
# Where was XXXX produced?

def type8(key):
    # keys = [type8, title]
    df = titles[titles["title"] == str(key[1])]
    production = list(df["production_countries"])
    years = list(df["release_year"])
    if len(production) == 1:
        production = [production[0]]
    else: 
        try:
            year = input("Which one are you asking about? Choose one of the following release years: " + str(years))
            index = years.index(int(year))
            production = [production[int(index)]]
            answer = "The production country of " + str(key[1].title()) + " is " + str(production[0]).upper() + "."
            return answer
        except: 
            print("Oops, the selected year is not found")
```

```python
# Type 9: Type
# What's XXX's type?
# Is XXX a show/movie?
# Is XXX a show or a movie?

def type9(key):
    # keys = [type9, title]
    df = titles[titles["title"] == str(key[1])]
    types = list(df["type"])
    years = list(df["release_year"])
    if len(types) == 1:
        types = [types[0]]
    else: 
        try:
            year = input("Which one are you asking about? Choose one of the following release years: " + str(years))
            index = years.index(int(year))
            types = [types[int(index)]]
            answer = str(key[1].title()) + " is a " + str(types[0]) + "."
            return answer
        except: 
            print("Oops, the selected year is not found")
```

```python
# Type 10: Vote
# How many people voted for XXX?

def type10(key):
    # keys = [type10, title]
    df = titles[titles["title"] == str(key[1])]
    votes = list(df["imdb_votes"])
    years = list(df["release_year"])
    if len(votes) == 1:
        votes = [votes[0]]
        if votes[0] == 0:
            return print("The number of people who rated this movie/show is unavailable.")
    else: 
        try:
            year = input("Which one are you asking about? Choose one of the following release years: " + str(years))
            index = years.index(int(year))
            votes = [votes[int(index)]]
            if votes[0] == 0:
                return print("The number of people who rated this movie/show is unavailable.")
            else:
                answer = str(int(votes[0])) + " people voted for " +str(key[1].title()) + "."
                return answer
        except: 
            print("Oops, the selected year is not found")
```
